**This code enables seamless integration with Streamlit, allowing users to input text files and receive a generated summary as output, with the number of lines in the summary being specified by the user.**

```python
import streamlit as st
from summariser import summary_generator

st.title("Text Summarization App")

def read_file(uploaded_file):
    if uploaded_file is not None:
        file_contents = uploaded_file.read().decode("utf-8")
        return file_contents
    return None

uploaded_file = st.file_uploader("Upload a text file", type=["txt"])
num_sentences = st.number_input("Enter the number of sentences for the summary:", min_value=1, step=1)

if uploaded_file is not None:
    file_contents = read_file(uploaded_file)
    if file_contents:
        title = uploaded_file.name.split(".")[0]  # use file name as title
        summary = summary_generator(file_contents, title, N=num_sentences)
        st.subheader("Generated Summary:")
        st.write(summary)
    else:
        st.warning("Please upload a text file.")
```
